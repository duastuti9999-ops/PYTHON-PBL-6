```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


data = pd.read_csv(r"urban_water_large_dataset.csv")
data.head()
```

```python

# Handle missing values
data.fillna(data.mean(numeric_only=True), inplace=True)
```

```python

# IQR Method for Outlier Removal
def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

data = remove_outliers(data)
```

```python

# Features & Target
X = data.drop("daily_liters_used", axis=1)
y = data["daily_liters_used"]
```

```python

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Train Model
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))

```

```python

# Graph 1: Actual vs Predicted
plt.figure()
plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```

```python

# Graph 2: Residual Plot
residuals = y_test - y_pred
plt.figure()
plt.scatter(y_pred, residuals)
plt.axhline(y=0)
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python

# Graph 3: Feature vs Target
plt.figure()
plt.scatter(data["household_size"], data["daily_liters_used"])
plt.xlabel("Household Size")
plt.ylabel("Water Usage")
plt.title("Household vs Usage")
plt.show()
```

```python

# Graph 4: Distribution
plt.figure()
plt.hist(data["daily_liters_used"], bins=30)
plt.xlabel("Water Usage")
plt.ylabel("Frequency")
plt.title("Distribution")
plt.show()
```
